```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import statsmodels.api as sm
import statsmodels.formula.api as smf 
```

```python
def get_unweighted_pmf(df: pd.DataFrame, column:str) -> pd.Series:
    freq = df[column].value_counts()
    pmf = freq / len(df)

    return pmf

def get_weighted_pmf(df: pd.DataFrame, column:str) -> pd.Series:

    total_pop = df['WTFA_C'].sum()
    group_df = df.groupby(by = column, observed= True)
    freq = group_df['WTFA_C'].sum()
    pmf = freq / total_pop

    return pmf
```

### CDF review

```python
def cal_base_cdf(srs: pd.Series) -> pd.Series:
    srtd_srs = np.sort(srs)
    
    cdf_values = np.arange(1, len(srtd_srs) + 1) / len(srtd_srs)
    
    return srtd_srs, cdf_values
```

```python
x, cdf = cal_base_cdf(df['AGEP_C'])
fig, ax = plt.subplots(figsize=(10, 5), layout='constrained')

ax.plot(x,cdf)
```

```python
ttl_x, ttl_cdf = cal_base_cdf(df['AGEP_C'])
ys_x, ys_cdf = cal_base_cdf(df[df['ADHDEV_C'] == 'Yes']['AGEP_C'])
no_x, no_cdf = cal_base_cdf(df[df['ADHDEV_C'] == 'No']['AGEP_C'])
```

```python
fig, ax = plt.subplots(figsize=(10, 5), layout='constrained')

ax.plot(ttl_x,ttl_cdf, label='Full Population')
ax.plot(ys_x,ys_cdf, label='ADHD')
ax.plot(no_x,no_cdf, label= 'Not ADHD')

ax.legend()
ax.set_xlabel('Age of the Child')
ax.set_ylabel('CDF')
ax.set_title('CDF\'s by Child Age Full Data Set')
```

# Hypothesis Testing

```python
full_pop = df['ADHDEV_C']
precovid = df[df['SRVY_YR']<2020]['ADHDEV_C']
postcovid = df[df['SRVY_YR']>=2020]['ADHDEV_C']
```

```python
full_pop_table = pd.crosstab(index=full_pop, columns = 'count')
precovid_table = pd.crosstab(index=precovid, columns = 'count')
postcovid_table = pd.crosstab(index=postcovid, columns = 'count')
```

```python
full_pop_ratios = full_pop_table/len(full_pop)
precovid_ratios = precovid_table/len(precovid)

precovid_expected = full_pop_ratios * len(precovid)
postcovid_full_expected = full_pop_ratios * len(postcovid)
postcovid_pre_expected = precovid_ratios * len(postcovid)
```

```python
print(f'Critical Value for 95% confidence {stats.chi2.ppf(q=0.95, df =3)}')
```

```python
print('Test there is no difference for pre-COVID19 ADHD values compared to the full population:')
print(stats.chisquare(f_obs=precovid_table, f_exp=precovid_expected))
```

We can accept the null hypothesis here as the p-value is less than 7.81

```python
print('Test there is no difference for post-COVID19 ADHD values compared to the full population:')
print(stats.chisquare(f_obs=postcovid_table, f_exp=postcovid_full_expected))
```

Again, we can accept that the p-value is less than 7.81

```python
print('Test There is no difference for post-COVID19 when Compared to pre-COVID19:')
stats.chisquare(f_obs=postcovid_table, f_exp=postcovid_pre_expected)
```

Here, we have to reject as the p-value is greater than 7.81.

### Regrestion Models

```python
newdf = df.reset_index()
newdf = newdf[['SRVY_YR','ADHDEV_C','HISPALLP_C','REGION','AGEP_C','WTFA_C']]
dummies_df = pd.get_dummies(newdf['ADHDEV_C'])
newdf = pd.concat([newdf,dummies_df], axis=1)
newdf['Yes'] = newdf['Yes'].astype(int)


dummies_df = None
newdf.head()
```

```python
precovid_df.HISPALLP_C.value_counts()
```

```python
precovid_df = newdf[newdf['SRVY_YR'] < 2020]
postcovid_df = newdf[newdf['SRVY_YR'] >= 2020]
```

```python
precovid_model = smf.logit('Yes ~  AGEP_C + REGION + HISPALLP_C', data=precovid_df)
precovid_results = precovid_model.fit()
```

```python
precovid_results.summary()
```

```python
postcovid_model = smf.logit('Yes ~ AGEP_C + REGION + C(HISPALLP_C)', data=postcovid_df)
postcovid_results = postcovid_model.fit()
```

```python
postcovid_results.summary()
```

```python
total_pop_model = smf.logit('Yes ~ AGEP_C + REGION + HISPALLP_C', data=newdf)
total_pop_results = total_pop_model.fit()
```

```python
total_pop_results.summary()
```

```python
def run_acc (model, results):
    endog = pd.DataFrame(model.endog, columns=[model.endog_names])
    exog = pd.DataFrame(model.exog, columns=model.exog_names)

    actual = endog['Yes']
    baseline = actual.mean()

    predict = (results.predict() >= 0.5)
    true_pos = predict * actual
    true_neg = (1 - predict) * (1 - actual)
    return (sum(true_pos) + sum(true_neg)) / len(actual)
```

```python
run_acc(precovid_model,precovid_results)
```

```python
run_acc(postcovid_model,postcovid_results)
```

```python
run_acc(total_pop_model,total_pop_results)
```
